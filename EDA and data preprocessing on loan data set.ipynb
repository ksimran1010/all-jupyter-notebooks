```python
def replaceMean(colName):
    colName= str(colName)
    dfAll[colName].fillna((dfAll[colName].mean()), inplace=True)
    print(dfAll[colName])
```

```python
def replaceZero(colName):
    colName= str(colName)
    dfAll[colName].fillna(0.0, inplace=True)
    print(dfAll[colName])
```

```python
def handledate(colName):
    dfAll[colName].fillna(' - ',inplace = True)
    new = dfAll[colName].str.split("-",n=2, expand = True) 
    dfAll[colName] = new[1]
    print(dfAll[colName])
    
```
